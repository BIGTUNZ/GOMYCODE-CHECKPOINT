```python
# Fetch Asteroid Data Using NEOWS API
neows_url = f"https://api.nasa.gov/neo/rest/v1/neo/browse?api_key={API_KEY}"
response = requests.get(neows_url)
asteroid_data = response.json()
```

```python
asteroids = asteroid_data["near_earth_objects"]
asteroids
```

```python
# Create an empty list to store the relevant asteroid data
asteroid_info = []


# Loop through the data to extract the required information
for asteroid in asteroids:
    asteroid_info.append({
        "Asteroid ID": asteroid["id"],
        "Asteroid name": asteroid["name"],
        "Minimal estimated diameter (km)": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
        "Absolute magnitude": asteroid["absolute_magnitude_h"],
        "Relative velocity (km/s)": asteroid["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"]
    })


#Convert this list into a pandas DataFrame
df_asteroids = pd.DataFrame(asteroid_info)
```
