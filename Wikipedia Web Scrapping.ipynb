```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
link = "https://en.wikipedia.org/wiki/2024_Summer_Olympics_medal_table#Medal_table"

page = requests.get(link)

soup = BeautifulSoup(page.text)
```

```python
title = soup.find('title')
```

```python
title
```

```python
article_content = {}
current_heading = None
for element in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p']):
    if element.name.startswith('h'):
                # Store the current heading
                current_heading = element.get_text().strip()
                article_content[current_heading] = []
    elif element.name == 'p' and current_heading:
                article_content[current_heading].append(element.get_text().strip())
```

```python
article_content
```

```python
wikipedia_links = []

for link in soup.find_all('a', href=True):
    href = link['href']
    if href.startswith('/wiki/') and not href.startswith('/wiki/Special:') and ':' not in href:
                full_url = f"https://en.wikipedia.org{href}"
                wikipedia_links.append(full_url)
wikipedia_links
```
